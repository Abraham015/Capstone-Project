```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from tqdm import tqdm
```

```python
def generate_data(**kwargs):
  raise NotImplementedError

def read_data(**kwargs):
  csv_filename = kwargs["csv_filename"]
  return pd.read_csv(csv_filename, index_col=0)
```

```python
ENV = "HW"
```

```python
get_data = read_data if ENV == "HW" else generate_data
```

```python
kwargs = {
    "csv_filename": "/content/DatosProyecto.csv"
}

data = get_data(**kwargs)

if not ENV == "HW":
  data.to_csv("DatosProyecto.csv")
```

```python
data.head()
```

```python
#Se usará 1 para femenino y 0 para masculino
data['Gender2'] = np.where(data['Gender'] == 1, 1, 0)
```

```python
#Para manejar un poco mejor los datos, se utlizará 1 para las personas que les toma más de 7 años terminar y 0 en caso contrario
data['HighTimeDegree'] = np.where(data['TimeDegree'] > "7", 1, 0)
```

```python
data['count'] = 1
```

```python
#Quitaremos las columnas que nos importan solamente
data2 = data[['HighTimeDegree','Gender2','count']]
data2.head()
```

```python
#Creamos un pivote para estas tablas
pd.pivot_table(
    data2, 
    values='count', 
    index=['HighTimeDegree'], 
    columns=['Gender2'], 
    aggfunc=np.size, 
    fill_value=0
)
```

```python
#Se utilizará 1 en caso de que se consiga más de 540 puntos en el examen, 0 en caso contrario
data['HighExamScoreGREV'] = np.where(data['GREV'] > 540, 1, 0)
#Se utilizará 1 en caso de que se consiga más de 691 puntos en el examen, 0 en caso contrario
data['HighExamScoreGREQ'] = np.where(data['GREQ'] > 691, 1, 0)
```
